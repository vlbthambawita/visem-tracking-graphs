```python
def visualize_graph_from_file(graph_path, title="Graph"):
    # Read the graph from the GraphML file
    graph = nx.read_graphml(graph_path)

    # Set node positions using a layout algorithm (you can change this to other layout algorithms)
    pos = nx.spring_layout(graph)

    # Draw nodes
    nx.draw_networkx_nodes(graph, pos, node_color="skyblue", node_size=500)

    # Draw node labels
    #nx.draw_networkx_labels(graph, pos, font_size=12, font_family="sans-serif")

    # Draw edges
    nx.draw_networkx_edges(graph, pos, edge_color="red", width=2)

    # Optionally, you can draw edge labels (weights) if they exist
    #edge_labels = nx.get_edge_attributes(graph, "weight")
    #if edge_labels:
    #    nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels, font_size=10)

    # Display the graph
    plt.axis("off")
    plt.title(title)
    plt.show()
```
